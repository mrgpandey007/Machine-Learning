# Matplotlib

```python
import matplotlib.pyplot as plt
l1=[1,2,4,9,25]    #x-axis
l2=[1,2,3,4,5]     #y-axis
plt.title("Log function")   #name/title of the graph
plt.plot(l1,l2)    #ploting the graph
#plt.show()         #used to show graph
plt.savefig('C:\\Machine Learning\\Matplotlib\\New folder.png')      #used to save image
plt.xlabel("Square of Number n")                 #used to name the x-axis
plt.ylabel("Number n")                 #used to name the y-axis
plt.show()
```

```python
plt.grid()
```

```python
plt.plot(l1,l2,color='cyan',linestyle='dashed',linewidth=2,marker='o',label='Hello World')
```

# Subplot()

```python
plt.subplot(2,2,1)
plt.plot(l2,l2)

plt.subplot(2,2,4)
plt.plot(l1,l2)

plt.subplot(2,2,2)
plt.plot(l2,l1)

plt.subplot(2,2,3)
plt.plot(l1,l1)
```

# Object oriented way

```python
fig=plt.figure()    # It creates the canva 
axis1=fig.add_axes([0.1,0.1,0.7,0.8])  ## setting the axes fig.add_axes([width_axis,height_axis,width,height])
axis1.plot(l1,l2)
axis2=fig.add_axes([0.1,0.1,0.5,0.4])  
axis2.plot(l1,l2)
```

```python
fig1=plt.figure()
axis3=fig1.add_axes([0.1,0.1,1,1])  
axis3.plot(l1,l2)
axis4=fig1.add_axes([0.3,0.6,0.3,0.3])  
axis4.plot(l1,l2)
```

```python
fig=plt.figure(figsize=(10,6),dpi=100)
axis1=fig.add_axes([0.1,0.1,0.7,0.8])  ## setting the axes fig.add_axes([width_axis,height_axis,width,height])
axis1.plot(l1,l2)
axis2=fig.add_axes([0.1,0.1,0.5,0.4])  
axis2.plot(l1,l2)
```

# Types of plots
