# Concat operation in data frame

```python
ar1=np.arange(0,25,1)
ar1.resize(5,5)
ar2=np.arange(25,50,1)
ar2.resize(5,5)

df7=pd.DataFrame(ar1)
df8=pd.DataFrame(ar2)
df_con=pd.concat([df7,df8])
df_con
```

```python
df_con=pd.concat([df7,df8],ignore_index=True) # the row index get unique 
df_con
```

```python
df_con={'Data1':df7,'Data2':df8}
pd.concat(df_con)
```

```python
df_con={'Data1':df7,'Data2':df8}
pd.concat(df_con,axis=1)
```

# Merge operation in Data Frame

```python
dict1={'ID':[101,102,103,104,105],
      'Samela Ram':[100000,0,50000,50000,0],
      'Jakhmola':[50000,40000,50000,50000,0]}
dict2={'ID':[106], 
       'Samela Ram':[200000],
       'Jakhmola':[190000]}
       
df9=pd.DataFrame(dict1)
df10=pd.DataFrame(dict2)
df_con1=pd.concat([df9,df10])

dict3={'ID':[101,102,103,104,105,106],
       'Renu Giri':[50000,50000,50000,0,50000,200000]}
df11=pd.DataFrame(dict3)
df_merge=pd.merge(df_con1,df11,on='ID')
df_merge
```

# Join operation in Data Frame

```python
emp_detail1={
    'id':[1,2,3,4,5],
    'Name1':['Saurav','Gaurav','Rajit','Isha','Neha'],
    'Branch_code':[101,101,102,102,101]}
emp_detail2={
    'id':[1,2,4,6,7],
    'Name2':['Rahul','Sazid','Wazid','Razid','Naina'],
    'Branch_code':[102,103,101,102,101]}

empdf1=pd.DataFrame(emp_detail1)
empdf2=pd.DataFrame(emp_detail2)
```

# 1.Full outer join (all joins are same as SQL concept)

```python
pd.merge(empdf1,empdf2,on='id',how='outer')
```

# 2.Inner Join

```python
pd.merge(empdf1,empdf2,on='id',how='inner')
```

# 3.Right Join

```python
pd.merge(empdf1,empdf2,on='id',how='right')
```

# 4.Left Join

```python
pd.merge(empdf1,empdf2,on='id',how='left')
```

# 5.Joining based on index

```python
pd.merge(empdf1,empdf2,right_index=True,left_index=True)
```
