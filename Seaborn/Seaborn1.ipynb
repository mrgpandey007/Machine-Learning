```python
sns.displot(df['total_bill'],bins=25) #histogram graph here bins are number of blue tower boxes
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=25)

plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=25)
```
