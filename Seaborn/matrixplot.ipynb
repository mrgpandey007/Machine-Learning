```python
import seaborn as sns
import matplotlib as plt
import numpy as np
import pandas as pd
```

# Heat Map

```python
f=sns.load_dataset('flights')
f
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tcor= tips[['total_bill','tip','size']]
tcor
```

```python
tcor.corr() #corr():-It is used to find corelation amongs the columns in the dataset/table
```

```python
sns.heatmap(tcor.corr())
```

```python
sns.heatmap(tcor.corr(),annot=True)
```

# Cluster Map

```python
sns.clustermap(tcor.corr())
```

# Pivot Table Heat map

```python
pvt=f.pivot_table(values='passengers',index='month',columns='year')
pvt
```

```python
sns.heatmap(pvt)
```
