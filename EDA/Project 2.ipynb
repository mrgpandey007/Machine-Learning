```python
def plot_(v,n):
    plt.subplot(2,2,n)
    sns.histplot(df[v],kde=True)

plot_('Age',1)
plot_('RestingBP',2)
plot_('Cholesterol',3)
plot_('MaxHR',4)
```

```python
df['Cholesterol'].value_counts()
```

```python
mean_ch = df.loc[df['Cholesterol']!=0,'Cholesterol'].mean()
mean_ch
```

```python
mean_ch=mean_ch.round(2)
df['Cholesterol']=df['Cholesterol'].replace(0,mean_ch)
```

```python
mean_re = df.loc[df['RestingBP']!=0,'RestingBP'].mean()
mean_re=mean_re.round(2)
df['RestingBP']=df['RestingBP'].replace(0,mean_re)
```
